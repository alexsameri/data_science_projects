# Python Pandas
Pandas is a fast, powerful, flexible and easy to use open source data analysis and manipulation tool, built on top of the Python programming language.                           
Pandas is an open-source library that is made mainly for working with relational or labeled data both easily and intuitively. It provides various data structures and operations for manipulating numerical data and time series.      

Pandas is a python library used for working with datasets. It has the function of analysing, cleaning, exploring and manipulating the data.         

##### Why Pandas?      
Pandas allows us to analyze big data and make conclusions based on statistical theories.                  
Pandas can clean messy data sets and make them readable and relevent, because relevent data is very important in data science. 

##### What can Pandas do?
Pandas gives information about the data like:              
Is there any corelation between two or more columns?    
What is the average value?        
What is the max value?           
What is the min value?              
We can also clean the data with pandas by deleting rows that are not relevent or contain wrong values like empty or null values.
                                              
There are two important data types defined by python:               
•	Series (Pandas series is like a column in a table or in is 1D array holding data of any type)                
•	DataFrame (Pandas DataFrame is a 2D data structure, array(2D), or table with rows and columns)

# Reading the data from files

Parameters:                             
•	Filename.txt: The name of filename and if required we have to mention the location of the file, if it is not in the same folder.                                                                                       
•	Sep: It is separator between the elements to identify to which column it belongs to.               
•	Header: If the file already has the first row as the column names, we don’t have to mention this parameter. By default, it will read the first line as the column names. •	Names: We will use ‘names =’ to mention the columns names which we want to create with.

# Subsetting data using Criteria

```python
df14 = pd.read_csv(r'surveys.csv')
df14.head()
```

```python
df14.shape
```

```python
df14.year.value_counts()
```

```python
df15 = df14[df14.year == 1977] # Subsetting only the values of year 2002
df15
```

```python
df16 = df14[df14.year != 2002] # Subsetting the values of year which is not 2002

df16
```

```python
df14[(df14.year >= 1980) & (df14.year <= 1985)] # subsetting values of year from 1980 to 1985
```

```python
# subsetting the the data frame with the required criteria and specifying the column name to be outputed

df14.loc[((df14.year >= 1980) & (df14.year <= 1985)),('record_id','year','sex','weight')] # (row,columns)
```

```python
df14.isnull().sum() # results to whether my value is null or not (True or False)
```

# Sorting the values

```python
df16 = pd.read_csv(r"Stores.csv")
df16
```

```python
df16.sort_values(by = 'Daily_Customer_Count', ascending = False) # Arranging the Daily_Customer_Count in ascending order
```

```python
# Arranging the Daily_Customer_Count in ascending order and outputing only first 5 rows by using head fucntion

df16.sort_values(by = 'Daily_Customer_Count', ascending = True).head()
```

```python
# Arranging the Daily_Customer_Count in ascending order and outputing only first 5 rows by slicing

df16.sort_values(by = 'Daily_Customer_Count', ascending = True)[:5]
```

```python
# Arranging the Daily_Customer_Count in ascending order and outputing values of Store_Sales which are greater than 60000

df16.sort_values(by = 'Daily_Customer_Count', ascending = True)[df16.Store_Sales > 60000]
```

```python
# Arranging the Daily_Customer_Count in ascending order; outputing values of Store_Sales which are greater than 60000;
# and selecting the required columns

df16.sort_values(by = 'Daily_Customer_Count', ascending = True)[["Store ID ","Daily_Customer_Count"]][df16.Store_Sales > 60000]
```

# Data Manipulation Operations

```python
# Creating a Data frame

df21=pd.DataFrame({"Country":["india","Australia","America","China","Russia"],
                   "income_in_crore":[600,700,800,900,1000],"Investment_in_crore":[6,4,5,6,7]})
df21
```

```python
# Creating a Data Frame

df22=pd.DataFrame({"Country":["india","London","Germany","Pakistan","Nepal"],
                   "temperature":[40,50,60,70,80],"Humidity":[10,20,30,40,50]})
df22
```

```python
# Concat function(one below other/vertically)

pd.concat((df21,df22),axis=0,ignore_index=False) # The table is combined one below the other vertically
```

```python
# Check the indexing after the first table and compare it with previous results

pd.concat((df21,df22),axis=0,ignore_index=True)
```

```python
# Combines the data set side by side
 
pd.concat((df21,df22),axis=1)  # horizontally.
```

```python
# combines the data based on country and shows only matching values

pd.merge(df21,df22,on = "Country") 
```

```python
#combines the data based on country and shows the matching and non matching values

pd.merge(df21,df22,on ="Country",how="outer")
```

```python
# combines the data based on country and shows non-matching values of first table and matching values of both the table

pd.merge(df21,df22,on = "Country",how="left")
```

```python
# combines the data based on country and shows non-matching values of second table and matching values of both the table

pd.merge(df21,df22,on = "Country",how="right")
```
