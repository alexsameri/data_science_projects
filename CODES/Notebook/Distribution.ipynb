# Distribution

### Normal Distribution

```python
# import the neccesary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
from scipy.stats import norm 
import seaborn as sns
```

```python
# Accessing the data
df = pd.read_csv('E:\Sankhyana Training/Employee data.csv')
df.head()
```

```python
# Calculating Mean
df['Age'].mean()
```

```python
# Calculating Standard deviation
df['Age'].std()
```

```python
# Create a normal distribution with mean as 42 and standard deviation as 10
mu = 42
std = 10
snd = stats.norm (mu, std)
snd
```

```python
# Generate 1000 random values between -100, 100
x = np.linspace(-100, 100, 1000)
x
```

```python
# Plot the standard normal distribution for different values of random variable falling in the range -100, 100
plt.figure(figsize=(7.5,7.5))
plt.plot(x, snd.pdf(x))
plt.xlim(-10, 100)
plt.title('Normal Distribution (Mean = 42, STD = 10)', fontsize='15')
plt.xlabel('Values of Random Variable X', fontsize='15')
plt.ylabel('Probability', fontsize='15')
plt.show()
```

## Bernoullis Distribution

```python
# importing the bernoulli library
from scipy.stats import bernoulli

X=[0,1] # ramdom variable
p=0.7   # probabiity of success
plt.figure(figsize=(7,7))
plt.xlim(-1, 2)
plt.bar(X, bernoulli.pmf(X,p),width=0.1,color=['lime','green'])
plt.title('Bernoulli Distribution (p=0.7)', fontsize='15')
plt.xlabel('Values of Random Variable X (0, 1)', fontsize='15')
plt.ylabel('Probability', fontsize='15')
plt.show()
```

# Binomial Distribution

```python
from scipy.stats import binom
import matplotlib.pyplot as plt
import numpy as np
#
# X = Discrete random variable representing number of successes
# p = Probability of the success
#
X = np.arange(0,21)
p = 0.6
n = 20
#
# Calculate binomial probability distribution
#
binom_pd = binom.pmf(X, n, p)
#
# Plot the probability distribution
#
fig, ax = plt.subplots(1, 1, figsize=(8, 6))
ax.plot(X, binom_pd ,'bo', ms=10)  # bo= blue circle markers and ms= marker size.
plt.ylabel("Probability", fontsize="18")
plt.xlabel("X - No. of Successes", fontsize="18")
plt.title("Binomial Distribution - No. of Successes Vs Probability", fontsize="18")
ax.vlines(X, 0,  binom_pd, colors='b', lw=5, alpha=0.5)
```

# Uniform Distribution

```python
uniform_data = stats.uniform.rvs(size=100000,  # Generate 100000 numbers
                                 loc = 0,      # From 0 
                                 scale=10)     # To 10
pd.DataFrame(uniform_data).plot(kind="density",  # Plot the distribution
                               figsize=(9,9),
                               xlim=(-1,11))
```

# Poisson Distribution

```python
from scipy.stats import poisson
#calculate probability
PEqual= poisson.pmf(k=5, mu=3)

print('The probability that the store sells 5 apples in a given day is =', PEqual)
```

```python
PD_less= poisson.cdf(k=4, mu=7)
print('The probability that the player kicks four or less goals in a given match is=', PD_less)
```

```python
PD_more= 1-poisson.cdf(k=20, mu=15)
print('The probability that the man who eats more than 20 cans of tuna in a given day is=',PD_more)
```

```python
#generate Poisson distribution with sample size 10000
x = poisson.rvs(mu=3, size=10000)

#create plot of Poisson distribution
plt.hist(x, density=True, edgecolor='black')
```

# Exponential distribution

```python
from scipy.stats import expon
#generate exponential distribution with sample size 10000
x = expon.rvs(scale=40, size=10000)

#create plot of exponential distribution
plt.hist(x, density=True, edgecolor='black')
```

```python
# calculate probability that x is less than 50 when mean rate is 40
ED= expon.cdf(x=50, scale=40)
print('The probability that well have to wait less than 50 minutes for the next eruption is ', ED)
```
