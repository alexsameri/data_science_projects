### Goal:        
Here our main goal is to create an artificial neural network to predict the promotion of the employee..


```python
#loading the data
import pandas as pd
df=pd.read_csv("E:\Sankhyana Training\Datasets\HR_analytics.csv")
df.head()
```

```python
#droping a column
data=df.drop(["Unnamed: 0"],axis=1)
data.head()
```

```python
#checking the data information
data.info()
```

```python
#checking for null values
data.isnull().sum()
```

```python
#spliting the data into target feature and input features
d=data.iloc[:,:12]
target=data.iloc[:,12]
```

```python
d
```

```python
target
```

```python
#spliting the data to training and testing sets

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(d,target,train_size=0.7)
```

### Building ANN 

```python
#install keras
%pip install keras
```

```python
#import neccesary libraries
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation,Dropout
```

### Creating hidden layers


```python
# Initialising the model(ANN)
model=Sequential()

# Add the input layer and the first hidden layer
model.add(Dense(units=12 , activation="relu"))
model.add(Dropout(0.2))

# Add the second hidden layer
model.add(Dense(units=24 , activation="relu"))
model.add(Dropout(0.2))

# Add output layer
model.add(Dense(units=1 , activation="sigmoid"))
```


                                                                                                                                            
### Train the ANN
    And
##### Compile the ANN 

```python
model.compile(optimizer="adam",loss="binary_crossentropy",metrics=["accuracy"])
```

### Fit the ANN 
  

```python
model.fit(X_train.values,Y_train.values,batch_size=1000,epochs=20)
```

##### Training of Artificial Neural Network

```python
loss,accuracy=model.evaluate(X_test.values,Y_test.values)
```

```python
import numpy as np
pred=model.predict(X_test.values)
x=[]
for i in np.squeeze(pred):
    x.append(np.rint(i))
predicted=pd.DataFrame(x)
print(pred> 0.5)
```

### checking the model accuracy                                 

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_test,predicted)
```
