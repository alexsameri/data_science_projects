## Data Visualization using python

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#generating a datafarme with help series
ds= pd.Series(np.random.randn(100), index =pd.date_range("1/1/2000", periods=100))

ds.tail()
```

```python
#ploting a line grapgh for the cummurative randomly gebnerated data
ds = ds.cumsum()
ds.plot();
```

```python
# Generating random value for the plots
df = pd.DataFrame(np.random.randn(100, 4), index=ds.index, columns=list("ABCD"))
df.head()
```

```python
df = df.cumsum()
df.plot()
```

```python
#plot a dotted line graph
x=[1, 2,3,4,5]
y=[6,7,8,9,10]

plt.plot(x, y, linewidth=6, linestyle='-.', color='b')

plt.xlabel("X_axis_label", fontdict={"fontsize": 18, "color": 'g'})
plt.ylabel("Y_axis_label", fontdict={"fontsize": 18, "color": 'g'})
plt.title("Dotted Line Graph", loc="center")
```

```python
#1st line graph x and y axis values
x=[1, 2,3,4,5]
y=[6,7,8,9,10]
plt.plot(x, y, linewidth=5)

#2nd line graph
x=[6, 7,8,9,10]
y=[1,2,3,4, 5]
plt.plot(x, y, linewidth=5)

plt.xlabel("X_axis_label", fontdict={"fontsize": 18, "color": 'g'}, labelpad=10)
plt.ylabel("Y_axis_label", fontdict={"fontsize": 18, "color": 'g'})
plt.title("Dotted Line Graph", loc="center")

plt.xlim((0,8))
plt.ylim((1,15))
```

```python
# Pie Chart

x_hrs = [1, 3, 8, 4]
daily_activities = ["Job","Swimming", "Sleeping", "Reading" ]

plt.pie(x_hrs, labels=daily_activities, autopct="%1.0f%%", explode=[0.1, 0.1, 0.2, 0], colors=['r', 'g', 'b', 'y'], startangle=70)
plt.show()
```

```python
#Creating a scatter plot

x =[1,2,3,4,5,3]
y= [6,7,8,9,10,6]
plt.scatter(x, y, c='r', marker=',')

x=[2,3,4]
y= [7,8,9]
plt.plot(x,y, c='y')

plt.show()
```

### Area plot / stack plot

```python

t_hrs= [1, 2,3,4,5]
sleeping = [8, 6,7,5,9]
works= [2,4, 6, 4, 3]

plt.stackplot(t_hrs, sleeping, works)
plt.show()
```

### Histogram

```python
df3= pd.DataFrame ({'a': np.random.rand(100)+1, 'b': np.random.rand(100), 'c': np.random.rand(100), 'd': np.random.rand(100)-1}, columns=['a','b','c','d'])
df3
```

```python
df3.plot.hist(alpha=0.5, bins=50, stacked=True)
```

### Subplots

```python
# working list
fig,ax=plt.subplots(nrows=2, ncols=2, figsize=[10,10])

histd= [9,8,7,5,4,8,3,8,3]
bins=[4,5,6,9]
bar_width=2
ax[0,0].hist(x=histd, bins=bins, width=bar_width)

x=[78,85, 75]
education= ['SSC','HSC', 'CS']
ax[0,1].bar(education,x)

x = [1,2,3,4,5,7]
y = [6,7,8,9,10,11]
ax [1, 0].scatter(x,y, s=100, c='r', marker=',')

x = [1,2,3,4,5,7]
y = [6,7,8,9,10,11]
ax[1,1].plot(x,y, c='r')

ax[0,0].set_title("Histogram")
ax[0,1].set_title("Bar")
ax[1,0].set_title("Scatter")
ax[1,1].set_title("Line Grapgh")
```
