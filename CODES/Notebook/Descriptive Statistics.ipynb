### Measure of Variance

```python
#Calculate the Rannge
mini = hr_data["Age"].min()
maxi = hr_data["Age"].max()

#print("The maximum value is:", maxi, "The min value is:", mini)
agerange=maxi-mini
print(agerange)
```

```python
# Variance
hr_data["Age"].var()
```

```python
#Staandard Deviation
hr_data["Age"].std()
```

```python
#Calculating the Interquatile Range

q1= hr_data["Age"].quantile(0.25)
q3 =hr_data["Age"].quantile(0.75)
iqr=q3-q1
iqr
```

### Percentile

```python
print("The 25th percentile is:", np.percentile(hr_data["Age"], 25))
print("The 50th percentile is:", np.percentile(hr_data["Age"], 50))
print("The 75th percentile is:", np.percentile(hr_data["Age"], 75))
print("The 90th percentile is:", np.percentile(hr_data["Age"], 90))
```
