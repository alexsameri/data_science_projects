```python
ames_df1 = ames_df.fillna(method="bfill")
```

```python
from sklearn.preprocessing import StandardScaler

le = StandardScaler()
```

```python
ames_df1['Age'] = ames_df1['Yr Sold'] - ames_df1['Year Built']
ames_df1['TotalSF'] = ames_df1['Total Bsmt SF'] + ames_df1['1st Flr SF'] + ames_df1['2nd Flr SF']
ames_df1['TotalBathrooms'] = ames_df1['Full Bath'] + 0.5 * ames_df1['Half Bath'] + ames_df1['Bsmt Full Bath'] + 0.5 * ames_df1['Bsmt Half Bath']
```

```python
ames_df2 = ames_df1[['Age', 'TotalSF', 'TotalBathrooms', 'SalePrice', 'Overall Cond']]
ames_df2.head()
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(ames_df2.Age, ames_df2.SalePrice)
plt.xlabel('Age')
plt.ylabel('SalePrice')
plt.title('Building age vs Salprice')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(ames_df2.TotalSF, ames_df2.SalePrice)
plt.xlabel('TotalSF')
plt.ylabel('SalePrice')
plt.title('Building TotalSF vs Salprice')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(ames_df2.TotalBathrooms, ames_df2.SalePrice)
plt.xlabel('TotalBathrooms')
plt.ylabel('SalePrice')
plt.title('Building TotalBathrooms vs Salprice')
plt.show()
```

```python
plt.figure(figsize=(10, 10))
plt.scatter(ames_df2['Overall Cond'], ames_df2.SalePrice)
plt.xlabel('Overall Cond')
plt.ylabel('SalePrice')
plt.title('Building Overall condition vs Salprice')
plt.show()
```

```python
train= ames_df2[: (int((len(ames_df2)*0.8)))]
test= ames_df2[(int((len(ames_df2)*0.8))):]
```

```python
train.shape
```

```python
test.shape
```

```python
from sklearn import linear_model
regr =  linear_model.LinearRegression()
train_y = np.array(train[['SalePrice']])
train_x = np.array(train[['TotalBathrooms', 'Overall Cond', 'TotalSF', 'Age']])
regr.fit(train_x, train_y)
```

```python
from sklearn.metrics import r2_score
test_x = np.array(test[['TotalBathrooms', 'Overall Cond', 'TotalSF', 'Age']])
test_y = np.array(test[['SalePrice']])
test_y_ = regr.predict(test_x)
print("Accuracy: %.2f" % r2_score(test_y_, test_y))
```
